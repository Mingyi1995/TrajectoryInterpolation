```python
import pandas as pd
from shapely.geometry import Point, LineString
```

```python
df = pd.read_csv('20200330_TripsEvenInterpolation.csv')
```

```python
df.head()
```

```python
sample = df[df['trip_id']==65583.0]
sample.head()
```

```python
lat = sample['LAT'].tolist()
lon = sample['LON'].tolist()
timeStamp = sample['BaseDateTime'].tolist()
```

```python
windowSize = 30
output = pd.DataFrame(columns=['StartTime','EndTime','line'])
# output.columns = ['time','LAT','LON']
for index in range(len(lat)-windowSize):
    latWindow = lat[index:index+30]
    lonWindow = lon[index:index+30]
    StartTime = timeStamp[index]
    EndTime = timeStamp[index+30]
    line = LineString([Point(i) for i in list(zip(latWindow,lonWindow))])
    temp = pd.DataFrame({'StartTime':[StartTime],'EndTime':[EndTime],'line':[line]})
    output = pd.concat([output,temp])
    index += 1
```
